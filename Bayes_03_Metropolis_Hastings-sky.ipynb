## Part A: MH Proposals and Acceptance Probabilities

> *Starting with a very general notation...*

Accepting draws from a ***proposal distribution*** 

$\tilde x^{(t)} \sim q(\tilde x^{(t)}|x^{(t-1)})$

according to a "coin flip"

$x^{(t)} = \left\{\begin{array}{ll}
\tilde x^{(t)} & \text{with probability } \min\left(1,\frac{p(\tilde x^{(t)})}{p(x^{(t-1)})}\frac{q(x^{(t-1)}|\tilde x^{(t)})}{q(\tilde x^{(t)}|x^{(t-1)})}\right) \quad \color{gray}{\begin{array}{c}\textrm{... is the normalizing constant needed for $p$?}\\\textrm{... and what happens if $q$ is symmetric?}  \end{array}}\\
x^{(t-1)} & \text{otherwise}
\end{array}\right.$

will produce draws from $p(x^{(t)})$.

### This is a two step conditional process<br>which altogether defines a <u>transition distribution</u>

$
\begin{align*}
&\quad\, \Pr(\textrm{accept}|\textrm{proposal},\textrm{current state}) q(\textrm{proposal}|\textrm{current state})\\ 
&={}
\Pr(\textrm{accept}|\textrm{proposal}) p(\textrm{proposal}|\textrm{current state})\\
&={} \Pr(x^{(t)} = \tilde x^{(t)}) q(\tilde x^{(t)}|x^{(t-1)}) = p(\tilde x^{(t)}|x^{(t-1)})
\end{align*}
$

so $q(\tilde x^{(t)}|x^{(t-1)})$ is our **proposal distribuiton** but $p(\tilde x^{(t)}|x^{(t-1)})$ is our actual **transition distribuiton**

### Does it work?

Let's see for 

$
\begin{align*}
p(x^{(t)}) & \rightarrow {} N(0,1)\\
q(\tilde x^{(t)}|x^{(t-1)}) & \rightarrow {} N(x^{(t-1)},\sigma)
\end{align*}
$


````python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

p_x_pdf = lambda x: stats.norm().pdf(x)
mh_tuning_parameter = 1.5
q_xt_given_xtm1 = lambda x: stats.norm(x, scale=mh_tuning_parameter)
proposal_distribution = q_xt_given_xtm1

m = 10000
x = np.zeros(m)
rejections = 0
for t in range(1,m):
    x_tilde = proposal_distribution(x[t-1]).rvs()
    acceptance_probability = min(1, p_x_pdf(x_tilde)/p_x_pdf(x[t-1]))
    if stats.uniform().rvs() < acceptance_probability:
        x[t] = x_tilde
    else:
        x[t] = x[t-1]
        rejections += 1

fig,ax = plt.subplots(1,2,figsize=(10,5))
ax[0].plot(x)
ax[1].hist(x, density=True)
x_support = np.linspace(-4,4,301)
ax[1].plot(x_support, p_x_pdf(x_support));
````

### Why does this work?

> *Changing and hinting with the the notation to begin being suggestive about our purposes...*

If we have a **stationary distribution** $p(\theta \color{gray}{|x})$ then **transition distribution (kernel)** $p$ (not to be confused with the **Metropolis-Hastings prposal distribution** $q$!) produces a sample from this **stationary distribution** if the ***Markov chain*** it produces is ***reversible***, meaning that

$\require{cancel}
\begin{align*}
p(\theta^{(t)}|\theta^{(t-1)})p(\theta^{(t-1)}\color{gray}{|x}) & = {}  p(\theta^{(t-1)}|\theta^{(t)})p(\theta^{(t)}\color{gray}{|x})\\\\
& \textrm{But is this true if we use a Metropolis-Hastings transition distribution (kernel)?}
\\\\
p(\theta^{(t)}\color{gray}{|x}) & \overset{MH}{=} {} \frac{\alpha_{\textrm{accept prob}}^{\textrm{forward}} q(  \theta^{(t)}| \theta^{(t-1)})}{\alpha_{\textrm{accept prob}}^{\textrm{backward}} q(  \theta^{(t-1)}|\theta^{(t)})}p(\theta^{(t-1)}\color{gray}{|x})\\
p( \theta^{(t)}\color{gray}{|x}) & = {} \frac{\min\left(1,\frac{p( \theta^{(t)}\color{gray}{|x})}{p(\theta^{(t-1)}\color{gray}{|x})}\frac{q(\theta^{(t-1)}| \theta^{(t)})}{q( \theta^{(t)}|\theta^{(t-1)})}\right) q( \theta^{(t)}|\theta^{(t-1)})}{\min\left(1,\frac{p( \theta^{(t-1)}\color{gray}{|x})}{p( \theta^{(t)}\color{gray}{|x})}\frac{q( \theta^{(t)}| \theta^{(t-1)})}{q( \theta^{(t-1)}| \theta^{(t)})}\right) q(  \theta^{(t-1)}|\theta^{(t)})}p(\theta^{(t-1)}\color{gray}{|x})\\\\
& \textrm{The fractions in the "$\min$" in the numerator or denomenator are reciprocals}\\
& \textrm{thus regardless of if the numerator fraction is equal to, greater than, or less than $1$}\\\\
p( \theta^{(t)}\color{gray}{|x}) & = {} \frac{p( \theta^{(t)}\color{gray}{|x})}{p(\theta^{(t-1)}\color{gray}{|x})}\frac{q(\theta^{(t-1)}| \theta^{(t)})}{q( \theta^{(t)}|\theta^{(t-1)})} \frac{q\theta^{(t)}|\theta^{(t-1)})}{q(\theta^{(t-1)}|\theta^{(t)})}p(\theta^{(t-1)}\color{gray}{|x}) \\\\
\xcancel{p( \theta^{(t)}\color{gray}{|x})} & = {} \frac{\xcancel{p(\theta^{(t)}\color{gray}{|x})}}{\xcancel{p(\theta^{(t-1)}\color{gray}{|x})}}\frac{\cancel{q(\theta^{(t-1)}| \theta^{(t)})}}{\cancel{q( \theta^{(t)}|\theta^{(t-1)})}} \frac{\cancel{q\theta^{(t)}|\theta^{(t-1)})}}{\cancel{q(\theta^{(t-1)}|\theta^{(t)})}}\xcancel{p(\theta^{(t-1)}\color{gray}{|x})}\\  \longrightarrow \quad 1 & ={} 1 \quad \textrm{ so the equation is true and reversibility holds}
\end{align*}$

This illustrates how the entire design and purpose of the **Metropolis-Hastings** algorithm specification (which seemingly appeared magically out of thin air) is in fact based on "reverse engineering" (if you will please excuse the pun) exactly what is needed to create a **reversible Markov Chain** (whose **stationary distribution** is the **target** of the **Metropolis-Hastings** algorithm).

## Part C: Back to Gibbs sampling

**Gibbs sampling** also creates a **reversible Markov Chain** whereby the samples, forwards or backwards, are from the same **stationary distribution** (which is the **joint posterior distribution** of the (random variable) parameters of the **full conditional (posterior) distributions**


````python
# loc is mean and scale is standard deviation
n = 30; x = stats.norm(loc=0, scale=1).rvs(size=n)
C,G = 2,10000  # chains, Gibbs samples
theta,tau = np.zeros([C,G]),np.zeros([C,G])
theta[:,0] = 1000  # initialization
for c in range(C):
    for g in range(1,G):
        # https://en.wikipedia.org/wiki/Gamma_distribution
        tau[c,g] = \
        stats.gamma(a=n/2+1, 
                    scale=2/((x-theta[c,g-1])**2).sum()).rvs()
        theta[c,g] = \
        stats.norm(x.mean(), 
                   scale=1/np.sqrt(tau[c,g]*n)).rvs()

fig,ax = plt.subplots(1, 4, figsize=(14,3)); ax[0].set_title("$\\theta$ Burn-In"); ax[1].set_title("$\\tau$ Burn-In"); ax[2].set_title("$\\theta$ Converged Markov Chain\nStationary Distribution Samples"); ax[3].set_title("$\\tau$ Converged Markov Chain\nStationary Distribution Samples")
burn = 20
demo = 120
for c in range(C):
    ax[0].plot(theta[c,:burn], label="$\\theta$ Chain "+str(c))
    ax[1].plot(tau[c,:burn], label="$\\tau$ Chain "+str(c))
    ax[2].plot(np.arange(burn, demo, dtype=int), theta[c,burn:demo], label="$\\theta$ Chain "+str(c))
    ax[3].plot(np.arange(burn, demo, dtype=int), tau[c,burn:demo], label="$\\tau$ Chain "+str(c))
ax[0].legend(); ax[1].legend(); ax[2].legend(); ax[3].legend();         
````

## Metropolis within Gibbs

1. All **full conditional distributions** are proportional to the **joint**.
2. A **Metropolis-Hastings** step can sample from a **full conditional** by proposing a sample and accepting it according to the MH acceptance probability.
3. The target density in the MH acceptance probability appears in the numerator and denomenator so it's normalizing constants cancel and all that's needed 

Accepting draws from a ***proposal distribution*** 

$\tilde \theta^{(t)} \sim q(\tilde \theta^{(t)}|\theta^{(t-1)})$

according to 

$\require{cancel}
\theta^{(t)} = \left\{\begin{array}{ll}
\tilde \theta^{(t)} & \text{with probability } \min\bigg(1,\frac{p(\tilde \theta^{(t)}|\tau^{(t)},\mathbf{x})}{p(\theta^{(t-1)}|\tau^{(t)},\mathbf{x})}\overset{\textrm{if symmetric}}{\cancel{\frac{q(\theta^{(t-1)}|\tilde \theta^{(t)})}{q(\tilde \theta^{(t)}|\theta^{(t-1)})}}^1}\bigg) \\
\theta^{(t-1)} & \text{otherwise}\\\\
\end{array}\right.$

will produce draws from $p(\theta^{(t)}|\tau^{(t)},\mathbf{x})$.

If we don't know the **full conditional distributions** *analytically* we can't use the *simple version* of **Gibbs sampling**. *But we do know the full conditional distributions proportionally up to their normalizing constants* and the unknown normalizing constants cancel in the target density ratios

$\frac{p(\tilde \theta^{(t)}|\tau^{(t)},\mathbf{x})}{p(\theta^{(t-1)}|\tau^{(t)},\mathbf{x})} = \frac{p(\tilde \theta^{(t)},\tau^{(t)},\mathbf{x})}{p(\theta^{(t-1)},\tau^{(t)},\mathbf{x})} = \frac{p(\mathbf{x}|\tilde \theta^{(t)},\tau^{(t)})p(\tilde \theta^{(t)})}{p(\mathbf{x}|\theta^{(t-1)},\tau^{(t)})p(\theta^{(t-1)})}$

so all we need is the **joint distribuiton** (as a function of the parameter of the **full conditional**, here $\theta$).

And an analogous **Metropolis within Gibbs** step exists for $\tau^{(t)}$.

````python
#q1

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# =======================
# ===== DATA SETUP =====
# =======================
np.random.seed(123)

n = 30           # Sample size
true_theta = 0   # True mean
true_sd = 1      # True standard deviation
data = stats.norm(loc=true_theta, scale=true_sd).rvs(size=n)

# =========================
# ===== MODEL PRIORS =====
# =========================
# Suppose a Normal prior for theta ~ Normal(theta0, tau0_sq)
theta0   = 0.0      # Prior mean
tau0_sq  = 1.0      # Prior variance
# Suppose a Gamma prior for tau ~ Gamma(alpha, lambda_) (shape-rate parameterization)
alpha    = 2.0
lambda_  = 1.0

# ================================
# ===== MCMC SAMPLER SETUP ======
# ================================
C = 2       # number of chains
G = 10000   # number of iterations per chain

theta = np.zeros((C, G))
tau   = np.zeros((C, G))

# Initialize
theta[:, 0] = 1000.0   # an arbitrary large initial value for theta
tau[:, 0]   = 1.0      # initial value for tau

# For the Metropolis step on tau, we use an *independent* proposal from Gamma(a=alpha, rate=lambda_)
proposal_shape = alpha
proposal_rate  = lambda_

# ===============================
# ===== LOG-JOINT FUNCTION =====
# ===============================
def log_joint_density(theta_, tau_):
    """
    Unnormalized log posterior = log(prior(theta_)) + log(prior(tau_)) + log(likelihood).
    
    1) Likelihood: data ~ Normal(theta_, 1/tau_). 
       log-likelihood = -0.5 * tau_ * sum((data - theta_)^2) + constant wrt data dimension
                        but we omit the constant w.r.t. data for simplicity.
    2) Prior(theta_): Normal(theta0, tau0_sq).
       log = -0.5*(theta_ - theta0)^2 / tau0_sq  (up to additive const)
    3) Prior(tau_): Gamma(shape=alpha, rate=lambda_).
       log = (alpha-1)*log(tau_) - lambda_*tau_  (up to additive const)
    """
    # If tau_ <= 0, we can return -inf to disallow negative tau_ (should not happen if we keep proposals positive).
    if tau_ <= 0:
        return -np.inf
    
    log_prior_theta = -0.5 * (theta_ - theta0) ** 2 / tau0_sq
    log_prior_tau   = (alpha - 1)*np.log(tau_) - lambda_*tau_
    log_likelihood  = -0.5 * tau_ * np.sum((data - theta_)**2)
    return log_prior_theta + log_prior_tau + log_likelihood


# ==================================================
# ===== METROPOLIS-within-GIBBS SAMPLING LOOP =====
# ==================================================
for c in range(C):  # for each chain
    for g in range(1, G):
        
        # ----------- (1) Gibbs update for THETA -----------
        # Because we know the closed-form conditional for theta in a normal-normal model:
        #   p(theta|tau, data) ~ Normal(mean_theta, var_theta).
        # That is a direct Gibbs step:
        curr_tau = tau[c, g-1]
        
        mean_theta = (curr_tau * np.sum(data) + theta0/tau0_sq) / (curr_tau*n + 1.0/tau0_sq)
        var_theta  = 1.0 / (curr_tau*n + 1.0/tau0_sq)
        
        # Draw new theta from that normal:
        new_theta = stats.norm.rvs(loc=mean_theta, scale=np.sqrt(var_theta))
        
        
        # ----------- (2) Metropolis update for TAU -----------
        # Propose a new tau from Gamma(proposal_shape, rate=proposal_rate)
        tau_proposed = stats.gamma.rvs(a=proposal_shape, scale=1.0/proposal_rate)
        
        # Current tau:
        old_tau = tau[c, g-1]
        
        # log-acceptance ratio for tau
        # (since proposal is independent of the old tau, the ratio of proposal densities
        #  is just log_q(old_tau) - log_q(new_tau) = 0 if the same Gamma is used for both
        #  old and new, but let's be explicit if we want to confirm.)
        
        log_numer = log_joint_density(new_theta, tau_proposed)
        log_denom = log_joint_density(new_theta, old_tau)
        
        # Because the proposal is independent, we also have:
        # log_q_tau(old_tau) - log_q_tau(tau_proposed), but let's define them:
        def log_q_tau(x):
            # log pdf of Gamma(a=alpha, rate=lambda_) = (alpha-1)*ln(x) - lambda_*x  (up to constant)
            return (proposal_shape - 1)*np.log(x) - proposal_rate*x
        
        # If you want the full expression:
        log_acc_ratio = (log_numer - log_denom) + (log_q_tau(old_tau) - log_q_tau(tau_proposed))
        
        # Accept/reject
        u = np.log(np.random.rand())  # uniform(0,1) in log scale
        if u < log_acc_ratio:
            new_tau = tau_proposed
        else:
            new_tau = old_tau
        
        # Store updated values
        theta[c, g] = new_theta
        tau[c, g]   = new_tau


# ============================
# ===== PLOTTING RESULTS =====
# ============================
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Unpack subplots
ax_theta_trace, ax_theta_hist = axes[0, 0], axes[0, 1]
ax_tau_trace,   ax_tau_hist   = axes[1, 0], axes[1, 1]

colors = ["blue", "orange"] 

for c in range(C):
    ax_theta_trace.plot(theta[c, :], alpha=0.6, color=colors[c], label=f"Chain {c+1}")
    ax_tau_trace.plot(tau[c, :], alpha=0.6, color=colors[c], label=f"Chain {c+1}")
    
    ax_theta_hist.hist(theta[c, :], bins=30, density=True, alpha=0.4, color=colors[c])
    ax_tau_hist.hist(tau[c, :],   bins=30, density=True, alpha=0.4, color=colors[c])

ax_theta_trace.set_title("Theta Trace Plot")
ax_tau_trace.set_title("Tau Trace Plot")
ax_theta_hist.set_title("Theta Posterior Distribution")
ax_tau_hist.set_title("Tau Posterior Distribution")

ax_theta_trace.legend()
ax_tau_trace.legend()
plt.tight_layout()
plt.show()
````

````python
#q2

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# =========================
# ===== DATA & SETUP =====
# =========================
np.random.seed(123)

n = 30
true_theta = 0.0
true_sd    = 1.0
data = stats.norm(loc=true_theta, scale=true_sd).rvs(size=n)

# Non-Normal prior for theta: Laplace(0, b=2), for example.
#   log p(theta) = -|theta| / b (up to additive constant)
def log_prior_theta_laplace(theta_val, b=2.0):
    return -np.abs(theta_val)/b

# Keep a Gamma prior for tau, Gamma(shape=alpha=2, rate=1).
alpha   = 2.0
lam     = 1.0
def log_prior_tau(tau_val):
    if tau_val <= 0:
        return -np.inf
    return (alpha - 1)*np.log(tau_val) - lam*tau_val

# Likelihood: data ~ Normal(theta, variance=1/tau).
# Unnormalized log-likelihood ~ -0.5 * tau * sum((data - theta)^2)
def log_likelihood(theta_val, tau_val, data_array):
    if tau_val <= 0:
        return -np.inf
    return -0.5 * tau_val * np.sum((data_array - theta_val)**2)

# Joint posterior log (up to constant)
def log_joint(theta_val, tau_val, data_array):
    return (log_prior_theta_laplace(theta_val)
            + log_prior_tau(tau_val)
            + log_likelihood(theta_val, tau_val, data_array))

# ---------------------------
# Dependent proposal for tau
# We'll use stats.truncnorm(a=-tau_current/s, b=inf, loc=tau_current, scale=s).
# We need the log of that proposal's pdf for acceptance ratio.
def log_truncnorm_pdf(x_from, x_to, scale):
    """
    Log pdf of TruncNorm around x_from with 'scale',
    truncated to (0, inf) => a = -x_from/scale, b = inf.
    """
    a_ = -x_from/scale
    b_ = np.inf
    return stats.truncnorm.logpdf(x_to, a=a_, b=b_, loc=x_from, scale=scale)

# ===============================
# ===== MCMC PARAMETERS ========
# ===============================
C = 2        # number of chains
G = 10000    # number of iterations per chain

theta = np.zeros((C, G))
tau   = np.zeros((C, G))

# Initialize
theta[:, 0] = 1.0   # can be any real, including outside prior's typical range
tau[:, 0]   = 1.0   # must be > 0

# ---------------------------
# Metropolis-within-Gibbs:
#   (1) Update theta by Metropolis
#   (2) Update tau by Metropolis with a dependent (truncnorm) proposal
# ---------------------------
proposal_sd_theta = 1.0   # for theta random-walk
proposal_sd_tau   = 0.5   # for tau truncated normal

for c in range(C):
    for g in range(1, G):
        old_theta = theta[c, g-1]
        old_tau   = tau[c, g-1]

        # ========== 1) Metropolis update for theta ==========
        # We'll do a simple random-walk Normal around old_theta.
        theta_prop = np.random.normal(loc=old_theta, scale=proposal_sd_theta)

        # Acceptance ratio in log-scale
        log_acc_theta = (
            log_joint(theta_prop, old_tau, data)
            - log_joint(old_theta, old_tau, data)
        )
        # random-walk is symmetric => no proposal ratio term needed
        if np.log(np.random.rand()) < log_acc_theta:
            new_theta = theta_prop
        else:
            new_theta = old_theta

        # ========== 2) Metropolis update for tau (dependent) ==========
        # Propose from truncated normal around old_tau
        a_ = -old_tau / proposal_sd_tau  # ensures positivity if old_tau>0
        b_ = np.inf
        trunc_dist = stats.truncnorm(a=a_, b=b_, loc=old_tau, scale=proposal_sd_tau)
        tau_prop   = trunc_dist.rvs()

        # Because the proposal is not symmetric, we need log_q_backward - log_q_forward
        log_q_forward  = log_truncnorm_pdf(old_tau, tau_prop, proposal_sd_tau)
        log_q_backward = log_truncnorm_pdf(tau_prop, old_tau, proposal_sd_tau)

        log_acc_tau = (
            log_joint(new_theta, tau_prop, data)
            - log_joint(new_theta, old_tau, data)
            + (log_q_backward - log_q_forward)
        )
        if np.log(np.random.rand()) < log_acc_tau:
            new_tau = tau_prop
        else:
            new_tau = old_tau

        # Store
        theta[c, g] = new_theta
        tau[c, g]   = new_tau

# ============================
#   PLOTTING THE RESULTS
# ============================
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
ax_theta_trace, ax_theta_hist = axes[0, 0], axes[0, 1]
ax_tau_trace,   ax_tau_hist   = axes[1, 0], axes[1, 1]

colors = ["blue", "orange"]
for c_ in range(C):
    # Trace plots
    ax_theta_trace.plot(theta[c_, :], alpha=0.5, color=colors[c_], label=f"Chain {c_+1}")
    ax_tau_trace.plot(tau[c_, :],   alpha=0.5, color=colors[c_], label=f"Chain {c_+1}")

    # Histograms
    ax_theta_hist.hist(theta[c_, :], bins=30, density=True, alpha=0.4, color=colors[c_])
    ax_tau_hist.hist(tau[c_, :],     bins=30, density=True, alpha=0.4, color=colors[c_])

ax_theta_trace.set_title("Theta Trace (Metropolis w/ Laplace prior)")
ax_tau_trace.set_title("Tau Trace (Truncnorm proposal)")
ax_theta_hist.set_title("Theta Posterior Distribution")
ax_tau_hist.set_title("Tau Posterior Distribution")

ax_theta_trace.legend()
ax_tau_trace.legend()
plt.tight_layout()
plt.show()
````

````python
from IPython.display import Markdown, display

# Put your LaTeX/Markdown content in a raw (r"") string:
markdown_content = r"""
# Q3

## Proposal Distribution

$$
q(\tilde{x}^{(t)} \mid x^{(t-1)})
= \left(\frac12\right)^{1-\tilde{x}^{(t)}}
  \left(\frac12\right)^{\tilde{x}^{(t)}}
$$

which means the proposal is symmetric:

$$
q(0 \mid 1) = q(1 \mid 0) = \tfrac12,
\quad
q(0 \mid 0) = q(1 \mid 1) = \tfrac12
$$

## Target (stationary) distribution

$$
p\bigl(x^{(t)}\bigr)
= \bigl(\tfrac13\bigr)^{\,1 - x^{(t)}}
  \bigl(\tfrac23\bigr)^{\,x^{(t)}}
$$

which implies

$$
p(0) = \tfrac13,
\quad
p(1) = \tfrac23
$$

## Metropolis--Hastings acceptance probability

$$
\alpha\bigl(x^{(t-1)} \rightarrow \tilde{x}^{(t)}\bigr)
= \min \Bigl(
  1,\,
  \frac{
    p\bigl(\tilde{x}^{(t)}\bigr)
  }{
    p\bigl(x^{(t-1)}\bigr)
  }
  \frac{
    q\bigl(x^{(t-1)} \mid \tilde{x}^{(t)}\bigr)
  }{
    q\bigl(\tilde{x}^{(t)} \mid x^{(t-1)}\bigr)
  }
\Bigr)
$$

Because the proposal distribution is symmetric,

$$
\frac{
  q\bigl(x^{(t-1)} \mid \tilde{x}^{(t)}\bigr)
}{
  q\bigl(\tilde{x}^{(t)} \mid x^{(t-1)}\bigr)
}
= 1
$$

so the acceptance probability simplifies to

$$
\alpha\bigl(x^{(t-1)} \rightarrow \tilde{x}^{(t)}\bigr)
= \min \Bigl(
  1,\,
  \frac{p(\tilde{x}^{(t)})}{p(x^{(t-1)})}
\Bigr)
$$

Substituting \(p(0)=\tfrac13\) and \(p(1)=\tfrac23\), we obtain:

$$
\alpha(1 \rightarrow 0)
= \min \Bigl(
  1,\,
  \frac{p(0)}{p(1)}
\Bigr)
= \min \Bigl(
  1,\,
  \frac{\tfrac13}{\tfrac23}
\Bigr)
= \tfrac12,
\quad
\alpha(0 \rightarrow 1)
= \min \Bigl(
  1,\,
  \frac{p(1)}{p(0)}
\Bigr)
= \min \Bigl(
  1,\,
  \frac{\tfrac23}{\tfrac13}
\Bigr)
= 1
$$

## Transition matrix

1. Probability of Staying at 0:

$$
K(0 \rightarrow 0)
= q(1 \mid 0)\bigl(1 - \alpha(0 \rightarrow 1)\bigr)
+ q(0 \mid 0)
= \tfrac12 (1-1) + \tfrac12
= 0.5
$$

2. Probability of Moving from 0 to 1:

$$
K(0 \rightarrow 1)
= q(1 \mid 0)\,\alpha(0 \rightarrow 1)
= \tfrac12 \times 1
= 0.5
$$

3. Probability of Moving from 1 to 0:

$$
K(1 \rightarrow 0)
= q(0 \mid 1)\,\alpha(1 \rightarrow 0)
= \tfrac12 \times \tfrac12
= 0.25
$$

4. Probability of Staying at 1:

$$
K(1 \rightarrow 1)
= q(0 \mid 1)\bigl(1 - \alpha(1 \rightarrow 0)\bigr)
+ q(1 \mid 1)
= \tfrac12 \bigl(1 - \tfrac12\bigr) + \tfrac12
= 0.75
$$

Thus, the transition kernel is:

$$
K
= \begin{pmatrix}
0.5 & 0.25 \\
0.5 & 0.75
\end{pmatrix}
$$

The stationary distribution \(\pi = (\pi_0, \pi_1)\) satisfies \(\pi K = \pi\).

Solving

$$
\pi_0 = 0.5\,\pi_0 + 0.25\,\pi_1,
\quad
\pi_1 = 0.5\,\pi_0 + 0.75\,\pi_1,
\quad
\pi_0 + \pi_1 = 1
$$

yields

$$
\pi_0 = \tfrac13,
\quad
\pi_1 = \tfrac23
$$

Hence, the stationary distribution is exactly as expected.
"""

# Display it
display(Markdown(markdown_content))
````
